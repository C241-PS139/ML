season pada waktu pembelian

```python
transaction_df["created_at"] = pd.to_datetime(transaction_df["created_at"])
```

```python
from datetime import datetime

transaction_df['bulan'] = transaction_df['created_at'].dt.strftime('%m')
```

```python
tran = transaction_df[['created_at', 'customer_id', 'payment_status',
                'total_amount', 'product_id', 'quantity','item_price','bulan']]
trans = tran[tran['payment_status'] == 'Success']
trans['product_id'] = trans['product_id'].astype(str)
trans.head()
```

```python
trans.info()
```

# Merge Dataset

```python
m1 = pd.merge(trans, cust, on='customer_id', how='left')
m1.head()
```

```python
print(m1.shape)
m1.info()
```

```python
m2 = pd.merge(m1,product,on="product_id",how='left')
m2
```

```python
print(m2.shape)
m2['product_id'].value_counts()
```

masterCategory khusus untuk Apparel

```python
m2_Ap = m2[m2['masterCategory'] == 'Apparel']
m2_Ap= m2.reset_index(drop=True)  # Mengatur ulang indeks tanpa menyimpan indeks asli
m2_Ap.head()
```

```python
data_filtered = m2_Ap.loc[~m2_Ap['season'].isin(['Winter'])]
data_filtered.head()
```

```python
data_1 = data_filtered.rename(columns={'season': 'temperature'})
data_1.head()
```

```python
data_1['temperature'] = data_1['temperature'].replace({
    'Summer': 'Hot',
    'Spring': 'Normal',
    'Fall': 'Cold'
})

data_1.head()
```

```python
# Menghapus duplikat berdasarkan 'productDisplayName'
data_deduplicated = data_1.drop_duplicates(subset=['productDisplayName'], keep='first')

# Menampilkan dataset yang telah dihapus duplikatnya
data_deduplicated.head()
```

```python
data_deduplicated["productDisplayName"].value_counts()
```

#Cleaning Merged Dataset

Memuat link menuju gambar

```python
image = pd.read_csv('images.csv')
image.head()
```

```python
data_deduplicated['product_id'].info()
```

```python
data_deduplicated['product_id'] = data_deduplicated['product_id'].astype(str)
data_deduplicated['product_id'] = data_deduplicated['product_id'].str.extract(r'(\d+)').astype(int)
data_deduplicated['product_id'].info()
```

```python
image = image.rename(columns={'filename': 'product_id'})
image['product_id'] = image['product_id'].astype(str)
image['product_id'] = image['product_id'].str.extract(r'(\d+)').astype(int)

m3 = pd.merge(data_deduplicated, image, on='product_id', how='left')
```

```python
m3.head()
```

```python
dataset = m3[['product_id','gender_product','temperature','usage','productDisplayName','city','link']]
dataset.head()
```

```python
dataset["city"] = dataset.apply(lambda trans: 'Jakarta' if trans['city'] == 'Jakarta Raya' else trans['city'], axis=1)
```

```python
dataset.to_csv('dataset.csv', index=False)
```
